```python
projects = pd.read_pickle('projects.zip')
projects.head(5)
```

```python
releases = pd.read_csv('releases.csv')
releases = releases.set_index(['project', 'name'])
releases.head(5)
```

Number of projects:

```python
releases.groupby(level=0).count().shape
```

Stars

```python
prj_lang_grp = projects.groupby('lang')
rl_lang_grp = releases.groupby('lang')
summary = pd.DataFrame(index=prj_lang_grp.count().index)
summary['mean stars'] = prj_lang_grp.mean()['stars']
summary['std stars'] = prj_lang_grp.std()['stars']
summary['total stars'] = prj_lang_grp.sum()['stars']
summary['total commits'] = rl_lang_grp.sum()['commits']
summary['total releases'] = releases.reset_index('name').groupby('lang').count()['name']
summary
```
